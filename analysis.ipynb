```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_paths = ['out_2024-03-04-19/out_2024-03-04-19-08-30_one_person_moving.csv',
              'out_2024-03-04-19/out_2024-03-04-19-09-45_standing.csv',
              'out_2024-03-04-19/out_2024-03-04-19-10-28_two_people_moving.csv',
              'out_2024-03-04-19/out_2024-03-04-19-11-38_one_person_standing_one_moving.csv']
```

### Time (x-axis) vs Y coordinate (y-axis) of ground truth, objectpose, personstate_prediction

```python
for file_path in data_paths:
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)

    # Converting stamp_secs to a readable datetime format for plotting
    # df["time"] = pd.to_datetime(df["stamp_secs"], unit='s')
    df['time'] = pd.to_datetime(df['stamp_secs'], unit='s') + pd.to_timedelta(df['stamp_nsecs'], unit='ns')

    # Plotting
    plt.figure(figsize=(12, 8))

    # Filtering and plotting for each topic
    for topic in df['topic'][~df['topic'].str.contains('velocity')].unique():
        subset = df[df["topic"] == topic]
        plt.plot(subset["time"], subset["positiony"], marker='.', linestyle='-', label=topic, alpha=0.7)

    plt.xlabel("Time")
    plt.ylabel("Y Coordinate")
    plt.title("Time vs Y Coordinate")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
